```python
import numpy as np
import pandas as pd
from pathlib import Path


# Load the dataset
data_path = Path("Helpdesk.csv")
helpdesk_df = pd.read_csv(data_path)

# helpdesk_df.head()
```

```python
# Drop any case with <2 events
helpdesk_df['Case ID'] = helpdesk_df['Case ID'].astype(str)
case_counts = helpdesk_df['Case ID'].value_counts()
valid_cases = case_counts[case_counts >= 2].index
helpdesk_df = helpdesk_df[helpdesk_df['Case ID'].isin(valid_cases)].reset_index(drop=True)

# Number of unique cases after filtering
print(helpdesk_df['Case ID'].nunique())

```
